## Clean CHAS DATA 2017-2021

```python
import pandas as pd
pd.options.mode.chained_assignment = None
```

```python
PATH = "datakind_housing_data/raw/050/Table8.csv"
raw_chas = pd.read_csv(PATH, encoding='latin-1')
```

```python
COLUMNS = [7, 20, 33, 46, 59, 10, 23, 36, 49, 62, 13, 26, 39, 52, 65, 2, 73, 86, 99, 112, 125, 76,  89, 102, 115, 128, 79, 92, 105, 118, 131, 68]
def create_columns(nums):
    outputs = []
    for n in nums:
        outputs.append('T8_est'+str(n))
    return outputs
raw_chas_cols = ['name','st', 'cnty'] + create_columns(COLUMNS)
raw_chas_cols
```

```python
raw_chas = raw_chas[raw_chas_cols]
```

```python
raw_chas.head()
```

```python
# chas = raw_chas[raw_chas['name'].str.contains('Florida')]
def get_county_name(s):
    if 'County' in s:
        splitted = s.split(' County')
    else:
        splitted = s.split(',')
    return splitted[0]

def get_state_name(s):
    if 'County' in s:
        splitted = s.split(' County')
    else:
        splitted = s.split(',')
    return splitted[1][1:]

raw_chas['county'] = raw_chas['name'].apply(get_county_name)
raw_chas['state'] = raw_chas['name'].apply(get_state_name)
raw_chas[['county', 'state']]
```

```python
raw_chas['state'].unique()
```

```python
def clean_state(s):
    if s[0] == ' ':
        s = s[1:]
    if s == 'District of Columbia':
        s = 'Maryland'
    return s
```

```python
raw_chas['state'] = raw_chas['state'].apply(clean_state)
len(raw_chas['state'].unique())
```

```python
chas = raw_chas
```

```python
owned_burdened_cols = [7, 20, 33, 46, 59]
owned_severe_burden_cols = [10, 23, 36, 49, 62]
rented_burdened_cols = [73, 86, 99, 112, 125]
rented_severe_burden_cols = [76,  89, 102, 115, 128]

chas['owned_burdened']= chas[create_columns(owned_burdened_cols)].sum(axis=1)
chas['owned_severe_burdend'] = chas[create_columns(owned_severe_burden_cols)].sum(axis=1)
chas['rented_burdened'] = chas[create_columns(rented_burdened_cols)].sum(axis=1)
chas['rented_severe_burdend'] = chas[create_columns(rented_severe_burden_cols)].sum(axis=1)
chas[['name', 'owned_burdened', 'owned_severe_burdend', 'rented_burdened', 'rented_severe_burdend']]
```

```python
len(chas['county'].unique())
```

```python
owned_nul_cols = [13, 26, 39, 52, 65]
rented_nul_cols = [79, 92, 105, 118, 131]
chas['owned_no_record']= chas[create_columns(owned_nul_cols)].sum(axis=1)
chas['rented_no_record'] = chas[create_columns(rented_nul_cols)].sum(axis=1)
chas['owned_population'] = chas['T8_est2'] - chas['owned_no_record']
chas['rented_population'] = chas['T8_est68'] - chas['rented_no_record']
chas[['name', 'owned_population', 'rented_population']]
```

```python
chas['owned_total_burdend'] = chas['owned_burdened'] + chas['owned_severe_burdend']
chas['rented_total_burdend'] = chas['rented_burdened'] + chas['rented_severe_burdend']
chas['owned_burdend_percentile'] = chas['owned_total_burdend'] / chas['owned_population']
chas['rented_burdend_percentile'] = chas['rented_total_burdend'] / chas['rented_population']
chas['owned_severe_burdend_percentile'] = chas['owned_severe_burdend'] / chas['owned_population']
chas['rented_severe_burdend_percentile'] = chas['rented_severe_burdend'] / chas['rented_population']
chas[['name', 'owned_burdend_percentile', 'owned_severe_burdend_percentile', 'rented_burdend_percentile', 'rented_severe_burdend_percentile']]
```

```python
chas.to_csv('datakind_housing_data/cleaned/chas_cost_burden.csv', index=False)
```
